##Model Development

```python
x = data.drop('Feeling depressed', axis =1 )
y = data['Feeling depressed']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report, accuracy_score
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
```

### Logistic Regression

```python

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

log_reg_grid = {"C": np.logspace(-4, 4, 20),
                "solver": ["liblinear"]}

# Setup grid hyperparameter search for LogisticRegression
gs_log_reg = GridSearchCV(LogisticRegression(),
                          param_grid=log_reg_grid,
                          cv=5,
                          verbose=True)

# Fit grid hyperparameter search model
gs_log_reg.fit(X_train, y_train);

# Check the best parameters
gs_log_reg.best_params_
```

```python
y_preds = gs_log_reg.predict(X_test)

accuracy_lg = accuracy_score(y_test, y_preds)
precision_lg = precision_score(y_test, y_preds)
recall_lg = recall_score(y_test, y_preds)
f1_lg = f1_score(y_test, y_preds)

print(f"Accuracy: {accuracy_lg}")
print(f"Precision: {precision_lg}")
print(f"Recall: {recall_lg}")
print(f"F1: {f1_lg}")
```

```python
from sklearn.metrics import confusion_matrix

def plot_conf_mat(y_test, y_preds, title):
    """
    Plots a confusion matrix using Seaborn's heatmap().
    """
    fig, ax = plt.subplots(figsize=(3, 3))
    ax = sns.heatmap(confusion_matrix(y_test, y_preds),
                     annot=True, # Annotate the boxes
                     cbar=False)
    plt.title(title)
    plt.xlabel("true label")
    plt.ylabel("predicted label")
```

```python
plot_conf_mat(y_test, y_preds, 'Logistic Regression')
```

### Decision Tree

```python
#feature Scaling
from sklearn.preprocessing import StandardScaler
st_x= StandardScaler()
x_train= st_x.fit_transform(X_train)
x_test= st_x.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier= DecisionTreeClassifier(criterion='entropy', random_state=0)
classifier.fit(x_train, y_train)
```

```python
y_pred_dt= classifier.predict(x_test)


accuracy_dt = accuracy_score(y_test, y_pred_dt)
precision_dt = precision_score(y_test, y_pred_dt)
recall_dt = recall_score(y_test, y_pred_dt)
f1_dt = f1_score(y_test, y_pred_dt)

print(f"Accuracy: {accuracy_dt}")
print(f"Precision: {precision_dt}")
print(f"Recall: {recall_dt}")
print(f"F1: {f1_dt}")
```
